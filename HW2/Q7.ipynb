```python
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
import seaborn as sns
```

1

```python
df = pd.read_csv("titanic.csv", sep=',')
df
```

```python
df.info()
```

```python
df = df.drop('Cabin', axis=1)
df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['Parch', 'SibSp','Ticket','Name','PassengerId'], axis=1, inplace=True)
df.info()
```

```python
df_back = df
```

```python
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=False, dtype=int)
```

```python
x = df.drop('Survived', axis=1)
y = df['Survived']
```

2

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

3

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier()
clf.fit(x_train, y_train)
accuracy_test = clf.score(x_test, y_test)
print(f'Accuracy on test is {accuracy_test*100:.2f}')
```

4

```python
from sklearn.preprocessing import StandardScaler
from copy import deepcopy

x_train_stand = deepcopy(x_train)
ss = StandardScaler()
x_train_stand = ss.fit_transform(x_train)
x_test_stand = ss.transform(x_test)
```

```python
clf = MLPClassifier()
clf.fit(x_train_stand, y_train)
accuracy_test = clf.score(x_test_stand, y_test)
print(f'Accuracy on test is {accuracy_test*100:.2f}')
```

استاندارد سازی در لحظه ای که من ران کردم باعث افزایش 72.9-76.64=3.74 درصد افزایش دقت بر روی داده های تست شد.

a

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score


MLP = MLPClassifier()
MLParameters = {
 'solver':['sgd', 'adam']   
}
MLPcv = GridSearchCV(MLP, param_grid=MLParameters)
MLPcv.fit(x_train_stand, y_train)
print(MLPcv.best_params_)
y_predict = MLPcv.best_estimator_.predict(x_test_stand)
print(accuracy_score(y_test,y_predict))
```

so Adam is better than SGD.
and we have 80.8-76.6=4.2 improvement

b

```python
from sklearn.model_selection import GridSearchCV

MLP = MLPClassifier()
MLParameters = {
 'learning_rate_init':[0.00001,0.0001,0.001,0.01,0.1]   
}
MLPcv = GridSearchCV(MLP, param_grid=MLParameters)
MLPcv.fit(x_train_stand, y_train)
print(MLPcv.best_params_)
y_predict = MLPcv.best_estimator_.predict(x_test_stand)
print(accuracy_score(y_test,y_predict))
```

So best learning rate init is 0.1.
and we have 82.2-76.6=5.6 improvement

c

```python
from sklearn.model_selection import GridSearchCV

MLP = MLPClassifier()
MLParameters = {
 'hidden_layer_sizes':[[100],[500],[1000],[100,100],[500,500],[1000,1000],[100,100,100],[500,500,500],[1000,1000,1000]],
}
MLPcv = GridSearchCV(MLP, param_grid=MLParameters)
MLPcv.fit(x_train_stand, y_train)
print(MLPcv.best_params_)
y_predict = MLPcv.best_estimator_.predict(x_test_stand)
print(accuracy_score(y_test,y_predict))
```

So best hidden layer count is 3 and best hidden layer neuron count per layer is 100.
and we have 79.4-76.6=2.8 improvement

d

```python
from sklearn.model_selection import GridSearchCV

MLP = MLPClassifier()
MLParameters = {
 'activation':['linear', 'tanh','relu']   
}
MLPcv = GridSearchCV(MLP, param_grid=MLParameters)
MLPcv.fit(x_train_stand, y_train)
print(MLPcv.best_params_)
y_predict = MLPcv.best_estimator_.predict(x_test_stand)
print(accuracy_score(y_test,y_predict))
```

So best activation is relu.
and we have 80.3-76.6=3.7 improvement

```python
from sklearn.model_selection import GridSearchCV

MLP = MLPClassifier()
MLParameters = {
 'solver':['sgd', 'adam'],
 'learning_rate_init':[0.00001,0.0001,0.001,0.01,0.1],
 'hidden_layer_sizes':[[100],[500],[1000],[100,100],[500,500],[1000,1000],[100,100,100],[500,500,500],[1000,1000,1000]],
 'activation':['linear', 'tanh','relu']   
}
MLPcv = GridSearchCV(MLP, param_grid=MLParameters)
MLPcv.fit(x_train_stand, y_train)
print(MLPcv.best_params_)
print(MLPcv.score(x_test_stand,y_test))
```

If we compute all them together we see that best activation is relu and best hidden layer size is [1000,1000,1000] and best learning rate init is 0.1 and best solver is sgd. and we have 78.9-76.6=2.3 improvement.
