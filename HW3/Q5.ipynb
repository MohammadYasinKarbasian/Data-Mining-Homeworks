# 2

```python
df = pd.get_dummies(df, columns=['thal' ,'pain type'], drop_first=False, dtype=int)
df
```

# 3

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X = pd.DataFrame(pca.fit_transform(df), columns=['Feature 1', 'Feature 2'])
X
```

# 4

```python
from sklearn.cluster import KMeans
sse = []
kmeans = []
for k in range(1, 21):
   temp = KMeans(n_clusters=k, random_state=1)
   temp.fit(X)
   kmeans.append(temp)
   sse.append(temp.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 21), sse)
plt.xticks(range(1, 21))
```

As you can see in this picture elbow happens in k=5.

```python
from kneed import KneeLocator
kl = KneeLocator(range(1, 21), sse, curve="convex", direction="decreasing")

print(f"Elbow is in K={kl.elbow}")
```

# 5

```python
from sklearn.metrics import silhouette_score

score = []

for i in range(2,21):
    label = kmeans[i-1].predict(X)
    score.append(silhouette_score(X, labels=label))
```

```python
import matplotlib.pyplot as plt
Ks = [i for i in range(2,21)]
plt.bar(Ks, score)
plt.show()
```

Best k based on silhouette_score is K=2 because have higher silhouette_score than others.

### Another simple way

```python
labels5 = kmeans[4].predict(X)
labels2 = kmeans[1].predict(X)
df2 = pd.DataFrame(X)
df2['Lable5'] = labels5
df2['Lable2'] = labels2
```

```python
l = list(df2['Lable5'].unique())
l.sort()
print("For K=5 we have:")
for i in l:
    temp = df2[df2['Lable5']==i]
    print("For cluster",i)
    print(temp.describe())
```

```python
l = list(df2['Lable2'].unique())
l.sort()
print("For K=2 we have:")
for i in l:
    temp = df2[df2['Lable2']==i]
    print("For cluster",i)
    print(temp.describe())
```
