```python
df_v2 = df.iloc[:,0:5]
for column_name in df_v2.columns:
    df_v2[column_name] = df_v2[column_name].fillna(df_v2[column_name].mean())
df_v2
```

## 10

We have several strategies as you can see what I have copy from sklearn manual:
"strategy : str, default='mean'
    The imputation strategy.

If "mean", then replace missing values using the mean along each column. Can only be used with numeric data.
If "median", then replace missing values using the median along each column. Can only be used with numeric data.
If "most_frequent", then replace missing using the most frequent value along each column. Can be used with strings or numeric data. If there is more than one such value, only the smallest is returned.
If "constant", then replace missing values with fill_value. Can be used with strings or numeric data."

```python
df_v4 = df.iloc[:, 0:5]
imput = SimpleImputer(strategy='median')
imput.fit(df_v4)
df_v4 = imput.transform(df_v4)
df_v4 = pd.DataFrame(df_v4, columns=df.columns[0:5])
df_v4
```

```python
df_v5 = df.iloc[:, 0:5]
imput = SimpleImputer(strategy='most_frequent')
imput.fit(df_v5)
df_v5 = imput.transform(df_v5)
df_v5 = pd.DataFrame(df_v5, columns=df.columns[0:5])
df_v5
```

```python
df_v6 = df.iloc[:, 0:5]
imput = SimpleImputer(strategy='constant', fill_value=-1)
imput.fit(df_v6)
df_v6 = imput.transform(df_v6)
df_v6 = pd.DataFrame(df_v6, columns=df.columns[0:5])
df_v6
```

KNN is one the best algorithms that handle missing values because we can use other features to find the nearest neighbors.<br>
Also Random forest can handle missing values by using surrogate splits, which are alternative splits that mimic the optimal split as closely as possible. Random forest can also be used for missing value imputation by using the proximity matrix, which measures the similarity between data points based on how often they end up in the same leaf node.
